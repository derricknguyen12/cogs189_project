```python
# Extract only valid EEG channel names
valid_eeg_channels = [ch for ch in data.columns if ch not in ["subject", "condition", "time_ms"]]
print(f"Valid EEG channels: {valid_eeg_channels}")
```

```python
conditions = data["condition"].unique()

epochs = {cond: {ch: [] for ch in valid_eeg_channels} for cond in conditions}

# Loop through each condition and extract time-series
for cond in conditions:
    cond_data = data[data["condition"] == cond]
    for ch in valid_eeg_channels:
        if ch in cond_data.columns and not cond_data[ch].isnull().all():
            epochs[cond][ch] = cond_data[ch].values
        else:
            print(f"Warning: Channel '{ch}' not found or contains only NaNs for condition {cond}. Skipping.")

print(f"Epoching completed! Extracted {len(conditions)} unique conditions.")

print(f"Unique conditions: {conditions[:3]}")
```

```python
# ERP Averaging and Data Saving Across Conditions and Channels
erp_avg = {}
for cond in conditions:
    erp_avg[cond] = {}
    for ch in valid_eeg_channels:
        if len(epochs[cond][ch]) > 0:
            erp_avg[cond][ch] = np.mean(epochs[cond][ch], axis=0)
        else:
            erp_avg[cond][ch] = np.nan

erp_df = pd.DataFrame(erp_avg)

time_values = np.linspace(data["time_ms"].min(), data["time_ms"].max(), len(erp_df))

erp_df["time_ms"] = time_values

erp_df.to_csv("erp_average_fixed.csv", index=False)
print("ERP averaging completed and saved!")
```

```python
# Performing T-Tests on ERP Values Between Conditions in Defined Time Window
from scipy.stats import ttest_ind

start_time = -1500
end_time = 300

# Extract ERP values in the defined time window for each condition
erp_values_range = {
    cond: erp_data[str(cond)].loc[
        (erp_data["time_ms"] >= start_time) & (erp_data["time_ms"] <= end_time)
    ].values for cond in conditions
}

for cond in conditions:
    print(f"Condition {cond} ERP values between {start_time}ms and {end_time}ms: {erp_values_range[cond]}")


valid_conditions = {cond: values for cond, values in erp_values_range.items() if len(values) > 1}

if len(valid_conditions) >= 2:
    # Perform T-test between all pairs of conditions
    for cond1 in valid_conditions:
        for cond2 in valid_conditions:
            if cond1 < cond2: # avoid repeating pairs
                t_stat, p_value = ttest_ind(valid_conditions[cond1], valid_conditions[cond2])
                print(f"T-test between Condition {cond1} and Condition {cond2}: t={t_stat:.3f}, p={p_value:.3f}")
else:
    print("Not enough data points for valid T-tests.")
```

- Condition 1 vs Condition 3:

```python
# Performing Mann-Whitney U test on ERP Values Between Conditions in Defined Time Window
start_time = -1500
end_time = 300

erp_values_range = {
    cond: erp_data[str(cond)].loc[
        (erp_data["time_ms"] >= start_time) & (erp_data["time_ms"] <= end_time)
    ].values for cond in conditions
}

for cond in conditions:
    print(f"Condition {cond} ERP values between {start_time}ms and {end_time}ms: {erp_values_range[cond]}")

from scipy.stats import mannwhitneyu

valid_conditions = {cond: values for cond, values in erp_values_range.items() if len(values) > 1}

if len(valid_conditions) >= 2:
    # Perform Mann-Whitney U test between all pairs of conditions
    for cond1 in valid_conditions:
        for cond2 in valid_conditions:
            if cond1 < cond2:  # Avoid repeating tests
                stat, p = mannwhitneyu(valid_conditions[cond1], valid_conditions[cond2])
                print(f"Mann-Whitney U test between Condition {cond1} and Condition {cond2}: stat={stat:.3f}, p={p:.3f}")
else:
    print("Not enough data points for valid Mann-Whitney U tests.")
```

```python
# T-Test Analysis of ERP Values Between Healthy and Schizophrenia Groups
import pandas as pd
from scipy.stats import ttest_ind

start_time = -1500
end_time = 300

healthy_indices = demographic[demographic[" group"] == 0].index
schizophrenia_indices = demographic[demographic[" group"] == 1].index

erp_data = pd.read_csv("button-tone-sz/ERPdata.csv")

# Extract ERP values within the time window for healthy and schizophrenia groups
erp_values_healthy = erp_data.iloc[healthy_indices]
erp_values_schizophrenia = erp_data.iloc[schizophrenia_indices]

erp_values_range_healthy = erp_values_healthy.loc[
    (erp_data["time_ms"] >= start_time) & (erp_data["time_ms"] <= end_time)
].values

erp_values_range_schizophrenia = erp_values_schizophrenia.loc[
    (erp_data["time_ms"] >= start_time) & (erp_data["time_ms"] <= end_time)
].values

# Perform a T-test between Healthy and Schizophrenia groups for the ERP values
if len(erp_values_range_healthy) > 1 and len(erp_values_range_schizophrenia) > 1:
    erp_values_range_healthy = erp_values_range_healthy.reshape(-1)
    erp_values_range_schizophrenia = erp_values_range_schizophrenia.reshape(-1)

    t_stat, p_value = ttest_ind(erp_values_range_healthy, erp_values_range_schizophrenia)
    print(f"T-test between Healthy and Schizophrenia: t={t_stat:.3f}, p={p_value:.3f}")
else:
    print("Not enough data points for valid T-test between Healthy and Schizophrenia.")

if p_value < 0.05:
    print("There is a significant difference in EEG signals between schizophrenia and healthy groups.")
else:
    print("No significant difference in EEG signals between schizophrenia and healthy groups.")
```

The output of the T-test between the Healthy and Schizophrenia groups shows a T-statistic of -0.045 and a p-value of 0.964. The T-statistic represents the difference between the means of the two groups (Healthy and Schizophrenia) in relation to the variability of the data. In this case, the T-statistic being close to 0 suggests that there is little difference between the means of the two groups in the ERP values within the defined time window. The p-value of 0.964 is much larger than the typical significance threshold of 0.05, indicating that the difference observed is not statistically significant.

This result suggests that there is no significant difference in the ERP values between the Healthy and Schizophrenia groups in the time window from -1500ms to 300ms. In other words, the ERP data in this specific time range does not provide strong evidence to distinguish between the two groups. Therefore, this T-test outcome can be interpreted as showing no meaningful difference in the ERP values between the groups for the selected time period in the context of the study.

```python
import numpy as np

# Compute Cohen's d for effect size
mean_healthy = np.mean(erp_values_range_healthy)
mean_schizophrenia = np.mean(erp_values_range_schizophrenia)
std_healthy = np.std(erp_values_range_healthy, ddof=1)
std_schizophrenia = np.std(erp_values_range_schizophrenia, ddof=1)

# Pooled standard deviation
pooled_std = np.sqrt(((len(erp_values_range_healthy) - 1) * std_healthy**2 + (len(erp_values_range_schizophrenia) - 1) * std_schizophrenia**2) /
                     (len(erp_values_range_healthy) + len(erp_values_range_schizophrenia) - 2))

# Cohen's d
cohens_d = (mean_healthy - mean_schizophrenia) / pooled_std
print(f"Cohen's d (Effect Size): {cohens_d:.3f}")
```

Despite the T-test showing no significant difference (with a p-value of 0.964) between the ERP values of the healthy and schizophrenia groups, the effect size (Cohen's d) further supports that there is almost no meaningful difference between the two groups in terms of the ERP values for the time window.

# **Relative voltage differences**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to calculate relative voltage difference
def compute_rvd(df):
    rvd_12 = abs(df[1] - df[2]) / ((abs(df[1]) + abs(df[2])) / 2) * 100
    rvd_13 = abs(df[1] - df[3]) / ((abs(df[1]) + abs(df[3])) / 2) * 100
    rvd_23 = abs(df[2] - df[3]) / ((abs(df[2]) + abs(df[3])) / 2) * 100

    # Average RVD across all three conditions
    rvd_overall = (rvd_12 + rvd_13 + rvd_23) / 3

    # Store results in a new DataFrame
    rvd_df = pd.DataFrame({
        'time_ms': df['time_ms'],
        'RVD_1_2': rvd_12,
        'RVD_1_3': rvd_13,
        'RVD_2_3': rvd_23,
        'RVD_Overall': rvd_overall
    })
    return rvd_df

# Compute RVD
rvd_df = compute_rvd(erp_df)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(rvd_df['time_ms'], rvd_df['RVD_1_2'], label='RVD 1-2', linestyle='dashed')
plt.plot(rvd_df['time_ms'], rvd_df['RVD_1_3'], label='RVD 1-3', linestyle='dotted')
plt.plot(rvd_df['time_ms'], rvd_df['RVD_2_3'], label='RVD 2-3', linestyle='dashdot')
plt.plot(rvd_df['time_ms'], rvd_df['RVD_Overall'], label='RVD Overall', linewidth=2)

plt.xlabel('Time (ms)')
plt.ylabel('Relative Voltage Difference (%)')
plt.title('Relative Voltage Difference Between Conditions')
plt.legend()
plt.grid(True)
```



*   High RVD 1-2 and 2-3 before 0 ms indicates strong differences in brain activity between these condition pairs prior to the event.
*   RVD 1-3 remains low throughout, showing similar EEG responses between these two conditions.
* All RVD values drop after 0 ms, suggesting brain activity converges post-stimulus or post-action.
* RVD Overall starts near 100% and steadily decreases, highlighting that most signal divergence occurs before the event.
* These patterns reveal temporal dynamics in EEG responses and can help identify critical time windows for distinguishing between conditions or groups.



#ANOVA

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols

#import data
erp_data = pd.read_csv("button-tone-sz/ERPdata.csv")

# Display the first few rows of the data
print(erp_data.head())

# Reshape Data
eeg_data_long = pd.melt(erp_data, id_vars=['condition'], var_name='electrode', value_name='voltage')

#Display Reshaped Data
print(eeg_data_long.head())
# Fit the model
model = ols('voltage ~ C(condition) + C(electrode) + C(condition):C(electrode)', data=eeg_data_long).fit()

# Perform the ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

## ANOVA Results

**C(condition)**
- **Significance:** p = 0.035

- **Interpretation:**

 - The significant effect of condition indicates that the EEG voltage (e.g., N100
 amplitude) varies significantly across the three conditions:
   1. **Button press with tone generation** (active condition).
   2. **Passive tone playback** (passive condition).
   3. **Button press without tone generation** (control condition).
 - This finding aligns with the hypothesis that sensory suppression (corollary discharge) differs between conditions. In healthy controls, the N100 amplitude is typically suppressed during active button presses with tone generation compared to passive playback. However, in schizophrenia patients, this suppression is often impaired, leading to less differentiation between conditions.
 - The significant effect of condition suggests that the task successfully elicits differences in neural responses, which are likely disrupted in schizophrenia.


**C(electrode)**
* **Significance:** p < 0.001

* **Interpretation:**
 - The highly significant effect of electrode indicates that EEG voltages vary significantly across the 9 electrode sites (Fz, FCz, Cz, FC3, FC4, C3, C4, CP3, CP4).

 - This is expected because different brain regions contribute differently to sensory processing and motor planning. For example:

   - **Frontal electrodes** (e.g., Fz, FCz): Associated with motor planning and corollary discharge.

   - **Central electrodes** (e.g., Cz, C3, C4): Associated with sensory processing and integration.

   - **Parietal electrodes** (e.g., CP3, CP4): Associated with higher-order sensory processing.

  - The variation across electrodes highlights the spatial specificity of neural responses, which is crucial for understanding how schizophrenia affects different brain regions.

**C(condition):C(electrode)** (Interaction)
 - **Significance:** p = 0.999

 - **Interpretation:**

  - The non-significant interaction suggests that the effect of condition on EEG voltage does not depend on the electrode site. In other words, the differences between conditions are consistent across all electrodes.

  - This finding implies that the neural mechanisms underlying sensory suppression (or lack thereof in schizophrenia) are broadly distributed across the brain rather than being localized to specific regions.

#SVM Model

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GroupShuffleSplit, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from imblearn.over_sampling import SMOTE

# Load ERP data & demographic data
erp_data = pd.read_csv("button-tone-sz/ERPdata.csv")
demographic = pd.read_csv("button-tone-sz/demographic.csv")
demographic.columns = demographic.columns.str.strip()

# Merge ERP data with demographic data based on subject ID
merged_data = erp_data.merge(demographic[['subject', 'group']], on='subject')

# Separate features (X) and labels (y)
X = merged_data.drop(['subject', 'condition', 'time_ms', 'group'], axis=1)  # EEG features
# Labels: 0 = Healthy Control (HC), 1 = Schizophrenia (SZ)
y = merged_data['group']

# Filter features to include only significant electrodes from ANOVA
significant_electrodes = ['Fz', 'Cz', 'FCz', 'C3', 'C4', 'CP3', 'CP4']
X_filtered = X[significant_electrodes]

# Add subject IDs to the filtered data
X_filtered.loc[:, 'subject'] = merged_data['subject'].values

# Feature Engineering
def extract_features(df):
    features = {
        'mean': df.mean(),
        'var': df.var(),
        'max': df.max(),
        'min': df.min(),
        'range': df.max() - df.min()
    }
    # Flatten the dictionary into a single row
    flattened_features = {}
    for feature_type, values in features.items():
        for electrode, value in values.items():
            flattened_features[f"{electrode}_{feature_type}"] = value
    return pd.Series(flattened_features)

# Apply feature engineering to each subject
X_aggregated = X_filtered.groupby('subject', group_keys=False).apply(extract_features).reset_index(drop=True)
y_aggregated = y.groupby(X_filtered['subject']).first()  # Labels are the same for all samples of a subject

# Check the shape of the aggregated data
print('--------------------------------------------------------------------')
print(f"Aggregated features shape: {X_aggregated.shape}", f"Aggregated labels shape: {y_aggregated.shape}\n")

# Split the data by subject (ensure no subject is in both training and test sets)
group_splitter = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=47)
train_idx, test_idx = next(group_splitter.split(X_aggregated, y_aggregated, groups=X_aggregated.index))

# Create training and test sets
X_train = X_aggregated.iloc[train_idx]
X_test = X_aggregated.iloc[test_idx]
y_train = y_aggregated.iloc[train_idx]
y_test = y_aggregated.iloc[test_idx]

# Check the distribution of labels in training and testing sets
print('--------------------------------------------------------------------')
print(f"Training labels distribution:\n{y_train.value_counts()}")
print(f"Testing labels distribution:\n{y_test.value_counts()}")

# Address class imbalance using SMOTE
smote = SMOTE(random_state=47)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Apply PCA to reduce dimensionality
pca = PCA(n_components=3)
X_train_pca = pca.fit_transform(X_train_resampled)
X_test_pca = pca.transform(X_test)

# Create an SVM pipeline with standardization and RBF kernel
svm_pipeline = make_pipeline(StandardScaler(), SVC(kernel='rbf', random_state=47, class_weight='balanced'))

# Parameter grid for GridSearchCV
param_grid = {
    'svc__C': [0.8, 0.9, 1, 1.1, 1.2, 2],  # Regularization parameter
    'svc__gamma': [0.1, 0.11, 0.12, 0.13, 0.14]  # Kernel coefficient
}

# Perform GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(svm_pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)
grid_search.fit(X_train_pca, y_train_resampled)

# Print the best parameters and accuracy
print('\n--------------------------------------------------------------------')
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best cross-validation accuracy: {grid_search.best_score_:.4f}\n")

# Evaluate the best model on the test set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test_pca)

# Print classification report and confusion matrix
print('--------------------------------------------------------------------')
print("Test Set Performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=['Healthy Control', 'Schizophrenia']))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print('\n--------------------------------------------------------------------')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mne
from mne import create_info
from mne.io import RawArray
from mne.channels import make_standard_montage
from mne.preprocessing import ICA

# Load the EEG dataset
erp_data = pd.read_csv("button-tone-sz/ERPdata.csv")

# Extract EEG channel names (excluding non-EEG columns)
eeg_channels = [ch for ch in erp_data.columns if ch not in ["subject", "condition", "time_ms"]]

# Create an MNE Info object with EEG channels
sfreq = 1000  # Adjust based on actual EEG sampling rate
info = create_info(ch_names=eeg_channels, sfreq=sfreq, ch_types="eeg")

# Apply the 10-20 Standard Montage
montage = make_standard_montage("standard_1020")
info.set_montage(montage, on_missing="warn")

# Convert EEG data into an MNE RawArray
eeg_data_for_raw = erp_data[eeg_channels].values.T
raw_data = RawArray(eeg_data_for_raw, info)

# Initialize and Fit ICA
ica = ICA(n_components=min(10, len(eeg_channels)), random_state=97)
ica.fit(raw_data)

# Plot ICA components
ica.plot_components()
plt.show()
```

### Analysis of ICA Output

Your ICA results show the **topographical maps** of the independent components extracted from the EEG signals. Below is an interpretation of the different ICA components:

#### **Key Observations**
- **ICA000:** Strong negative activation over the entire scalp, possibly reflecting a global artifact or a neural source with widespread influence.
- **ICA001:** A more centralized component with low amplitude, possibly reflecting background brain activity.
- **ICA002 & ICA004:** Strong frontal dominance, likely representing **eye-blink artifacts**.
  - Blinks typically appear as **positive activation in the front** and **negative in the back**.
  - These components should be reviewed and possibly removed.
- **ICA003:** Shows lateralized activity across hemispheres, which could relate to **horizontal eye movements** or **frontal brain activity**.
- **ICA005 & ICA006:** These components display more **distributed patterns** with no clear dominance, potentially neural or mixed artifacts.
- **ICA007:** A focal activation at the center, possibly **muscle activity** from head or facial tension.
- **ICA008:** Displays high-amplitude localized activation in the **parietal region**, which could be related to **parietal brain processes** or **residual muscle artifacts**.

#### **Potential Artifacts & Recommendations**
- **Eye Blink & Eye Movements:** ICA002, ICA004 → Likely candidates for removal.
- **Muscle Artifacts:** ICA007, ICA008 → Check frequency content before removal.
- **Neural Components:** ICA000, ICA003, ICA005 → Likely brain activity, should be preserved.
